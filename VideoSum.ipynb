```python
from art import *
import colorama
import pytube as p
import pytube.exceptions
import os
from transformers import pipeline
from pytube import YouTube
from pytube import extract

class ChoiceException(Exception):  # Custom Exception class
    def __init__(self, value):
        self.value = value


print(colorama.Fore.YELLOW)
tprint("VideoSum")
print(colorama.Fore.RESET)
while True:
    try:
        input_link = input("⌨ Enter a valid YouTube Video Link: ")
        yt = p.YouTube(input_link)  # Create a YouTube object for the video with the given link
        print("⌛ We are working on generating your transcript... Please wait...")
        
        # Get the audio stream
        audio = yt.streams.filter(only_audio=True).first()
        
        # Download the audio in MP4 format
        audio_path = audio.download()
        
        # Convert the MP4 audio to WAV format using ffmpeg
        wav_filename = os.path.splitext(audio_path)[0] + '.wav'
        os.system(f'ffmpeg -i "{audio_path}" "{wav_filename}"')
        
        print("✅ TRANSCRIPT GENERATED SUCCESS!")
        break
    except pytube.exceptions.RegexMatchError:  # Handles all the input exceptions
        print("🛑 Invalid URL! 🛑")


cls = pipeline("automatic-speech-recognition")
result = cls(wav_filename)

print("--------------------------------------------------------")
print("--------------------------------------------------------")
print(colorama.Fore.GREEN)
print("Summary: ")
print(result)
print(colorama.Fore.RESET)
print(colorama.Fore.MAGENTA)
tprint("========================")
tprint("CREDS:        MINSIC        WARRIORS")
tprint("========================")
print(colorama.Fore.RESET)
```
